```python
df['Grad.Rate']['Cazenovia College'] = 100

# Try removing various columns (features) from the dataset and examin if it improves/degrades your K-Means model performance, or it may have little impact.
# Report 10 cases where you removed one or more features and indicate how it impacted the model performance.
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score



# Prepare data for clustering
X = df.select_dtypes(include=[np.number])  # Select only numerical features

# Initial K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
initial_labels = kmeans.labels_

# Calculate sse and silhouette score
initial_silhouette = silhouette_score(X, initial_labels)
print("Inertia (SSE):", kmeans.inertia_)
print(f'Initial Silhouette Score: {initial_silhouette}')
```

```python
features_to_remove = [
    ['apps'],                    # Case 1
    ['apps', 'f_undergrad'],              # Case 2
    ['apps', 'f_undergrad', 'p_undergrad'],     # Case 3
    ['apps', 'f_undergrad', 'p_undergrad', 'outstate'],   # Case 4
    ['apps', 'f_undergrad', 'p_undergrad', 'outstate', 'room_board'],  # Case 5
    ['apps', 'f_undergrad', 'p_undergrad', 'outstate', 'room_board','books'],     # Case 6
    ['apps', 'f_undergrad', 'p_undergrad', 'outstate', 'room_board','books', 'personal'] ,         # Case 7
    ['apps', 'f_undergrad', 'p_undergrad', 'outstate', 'room_board','books', 'personal',  'terminal' ] ,      # Case 8
    ['apps', 'f_undergrad', 'p_undergrad', 'outstate', 'room_board','books', 'personal',  'terminal', 's_f_ratio' ] ,    # Case 9
    ['apps', 'f_undergrad', 'p_undergrad', 'outstate', 'room_board','books', 'personal',  'terminal', 's_f_ratio','perc_alumni' ]   # Case 10
]


results = []

for features in features_to_remove:
    X_reduced = X.drop(columns=features)
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans.fit(X_reduced)
    labels = kmeans.labels_
    silhouette = silhouette_score(X_reduced, labels)
    results.append((features, silhouette))

# Print results
for features, score in results:
    print(f'Removed features: {features}, Silhouette Score: {score}')
```

```python
#The results indicate that removing the feature apps consistently leads to lower silhouette scores, suggesting it may be significant for clustering. 
#However, as more features are removed—specifically f_undergrad, p_undergrad, and others—the silhouette score improves, peaking at 0.5302 when nine features are excluded. 
#This suggests that certain combinations of features enhance clustering performance, highlighting the importance of careful feature selection.
```
