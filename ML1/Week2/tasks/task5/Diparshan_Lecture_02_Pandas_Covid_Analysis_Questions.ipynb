# Task: Covid-19 Data Analysis
### This notebook is used to understand the comprehension of Data Analysis techniques using Pandas library.

### Data Source: 
https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_daily_reports

### File naming convention

MM-DD-YYYY.csv in UTC.

### Field description

- Province_State: China - province name; US/Canada/Australia/ - city name, state/province name; Others - name of the event (e.g., "Diamond Princess" cruise ship); other countries - blank.

- Country_Region: country/region name conforming to WHO (will be updated).

- Last_Update: MM/DD/YYYY HH:mm (24 hour format, in UTC).

- Confirmed: the number of confirmed cases. For Hubei Province: from Feb 13 (GMT +8), we report both clinically diagnosed and lab-confirmed cases. For lab-confirmed cases only (Before Feb 17), please refer to who_covid_19_situation_reports. For Italy, diagnosis standard might be changed since Feb 27 to "slow the growth of new case numbers." (Source)

- Deaths: the number of deaths.

- Recovered: the number of recovered cases.

### Question 1

#### Read the dataset

```python
import pandas as pd

ds = pd.read_csv('07-17-2020.csv')
ds.head()
```

#### Display the top 5 rows in the data

```python
ds.head(5)
```

#### Show the information of the dataset

```python
print(type(ds))
ds.columns
```

#### Show the sum of missing values of features in the dataset

```python
missing_values = ds.isnull().sum()
print( missing_values)
```

### Question 2

#### Show the number of Confirmed cases by Country

```python
confirmed_cases = ds.groupby('Country_Region')['Confirmed'].sum()
print(confirmed_cases)
```

#### Show the number of Deaths by Country

```python
deaths_count = ds.groupby('Country_Region')['Deaths'].sum()
print(deaths_count)
```

#### Show the number of Recovered cases by Country

```python
recovered_cases = ds.groupby('Country_Region')['Recovered'].sum()
print(recovered_cases)
```

#### Show the number of Active Cases by Country

```python
active_cases = ds.groupby('Country_Region')['Active'].sum()
print(active_cases)
```

#### Show the latest number of Confirmed, Deaths, Recovered and Active cases Country-wise

```python
country_wise = pd.pivot_table(ds,
                              values=['Confirmed', 'Deaths', 'Recovered', 'Active'],
                              index=['Country_Region'],
                              aggfunc='sum').reset_index()
print(country_wise)
                    
```

### Question 3

### Show the countries with no recovered cases

```python
no_recovered = ds[ds['Recovered'] == 0]
no_recovered_result = no_recovered.groupby('Country_Region')['Recovered'].sum()

print(no_recovered_result)
```

#### Show the countries with no confirmed cases

```python
no_confirmed = ds[ds['Confirmed'] == 0]
no_confirmed_result = no_confirmed.groupby('Country_Region')['Confirmed'].sum()
print(no_confirmed_result)
```

#### Show the countries with no deaths

```python
no_deaths = ds[ds['Deaths'] == 0]
no_deaths_result = no_deaths.groupby('Country_Region')['Deaths'].sum()
print(no_deaths_result)
```

### Question 4

#### Show the Top 10 countries with Confirmed cases

```python
confirmed_cases = ds.groupby('Country_Region')['Confirmed'].sum().sort_values(ascending=False).head(10)
print(confirmed_cases)
```

#### Show the Top 10 Countries with Active cases

```python
active_cases = ds.groupby('Country_Region')['Active'].sum().sort_values(ascending=False).head(10)
print(active_cases)
```

### Question 5

#### Plot Country-wise Total deaths, confirmed, recovered and active casaes where total deaths have exceeded 50,000

```python
import matplotlib.pyplot as plt
```

```python
data = ds.groupby('Country_Region')[['Deaths', 'Confirmed', 'Recovered', 'Active']].sum()
filtered_data = data[data['Deaths'] > 50000]
filtered_data = filtered_data.sort_values(by='Deaths', ascending=False)

plt.figure(figsize=[12,6])

plt.plot(filtered_data.index, filtered_data['Deaths'], color='blue', label='Deaths')
plt.plot(filtered_data.index, filtered_data['Confirmed'], color='green', label='Confirmed')
plt.plot(filtered_data.index, filtered_data['Recovered'], color='red', label='Recovered')
plt.plot(filtered_data.index, filtered_data['Active'], color='black', label='Active')

plt.title('Total deaths > 50,000, confirmed, recovered and active casaes')
plt.xlabel('Country')
plt.ylabel('Cases')

plt.legend()
plt.tight_layout()
plt.show()
```

### Question 7

### Plot Province/State Wise Active Cases in USA

```python
country_data = ds[ds['Country_Region'] == 'US']
active_cases = country_data.groupby('Province_State')['Active'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
active_cases.plot(kind='bar', color='blue', edgecolor='black')
plt.title('State wise Active cases reported of COVID-19 in USA')
plt.xlabel('Province/State')
plt.ylabel('Active Cases')
plt.tight_layout()

plt.show()
```

### Question 8

### Plot Province/State Wise Confirmed cases in USA

```python
country_data = ds[ds['Country_Region'] == 'US']
confirmed_cases = country_data.groupby('Province_State')['Confirmed'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
confirmed_cases.plot(kind='bar', color='blue', edgecolor='black')
plt.title('State wise Confirmed Cases reported of COVID-19 in USA')
plt.xlabel('Province/State')
plt.ylabel('Confirmed Cases')
plt.tight_layout()

plt.show()
```

### Question 9

### Plot Worldwide Confirmed Cases over time

```python
import plotly.express as px
import plotly.io as pio
```

```python
ds['Last_Update'] = pd.to_datetime(ds['Last_Update'])

confirmed_worldwide = ds.groupby('Last_Update')['Confirmed'].sum().reset_index()

fig = px.line(confirmed_worldwide,
              x='Last_Update',
              y='Confirmed',
              title='Worldwide Confirmed Novel Coronavirus Cases Over Time',
              labels={'Last_Update': 'Date', 'Confirmed': 'Confirmed Cases'})

fig.update_traces(line_color='green')

fig.show()
```
